```python
%load_ext sql
%config SqlMagic.displaylimit = 50
%config SqlMagic.displaycon = False
%sql sqlite:///dwellings.db  
import sqlite3
```

```python
conn = sqlite3.connect('dwellings.db')
df.to_sql('rooms', conn, if_exists='replace', index=False)
```

```python
%%sql
SELECT layout_area_type, noise_traffic_day
FROM rooms
WHERE noise_traffic_day < 20
```

```python
%%sql
SELECT layout_area_type, sun_201806211200_mean
FROM rooms
ORDER BY sun_201806211200_mean DESC
LIMIT 10
```

```python
%%sql
SELECT layout_area_type, COUNT(*) AS room_count
FROM rooms
GROUP BY layout_area_type
```

```python
%%sql
SELECT layout_area_type,
       MAX(sun_201806211200_mean) AS best,
       MIN(sun_201806211200_mean) AS worst
FROM rooms
GROUP BY layout_area_type
```

```python
%%sql
SELECT layout_area_type, AVG(noise_traffic_day) AS avg_noise
FROM rooms
GROUP BY layout_area_type
HAVING avg_noise < 50
-- useless dataset
```

```python
%%sql
SELECT layout_area_type,
       AVG(sun_201806211200_mean) AS summer,
       AVG(sun_201812211200_mean) AS winter
FROM rooms
GROUP BY layout_area_type
```

```python
%%sql
SELECT layout_area_type,
       COUNT(*) AS room_count,
       AVG(layout_area) AS avg_area,
       AVG(sun_201806211200_mean) AS avg_sun,
       AVG(noise_traffic_day) AS avg_noise
FROM rooms
GROUP BY layout_area_type
ORDER BY avg_sun DESC
```

```python
%%sql
SELECT layout_area_type,
       AVG(sun_201806211200_mean) AS summer_noon,
       AVG(sun_201812211200_mean) AS winter_noon
FROM rooms
GROUP BY layout_area_type
ORDER BY summer_noon DESC
```

```python
%%sql
SELECT layout_area_type,
       AVG(sun_201806211200_mean) AS avg_sun,
       AVG(noise_traffic_day) AS avg_noise
FROM rooms
GROUP BY layout_area_type
HAVING avg_sun > (SELECT AVG(sun_201806211200_mean) FROM rooms)
   AND avg_noise < (SELECT AVG(noise_traffic_day) FROM rooms)
```

```python
result = pd.read_sql_query("SELECT layout_area_type, AVG(sun_201806211200_mean) AS avg_sun FROM rooms GROUP BY layout_area_type", conn)
result.plot(kind='bar', x='layout_area_type', y='avg_sun')
plt.show()
```

```python
%%sql
SELECT layout_area_type,
       AVG(layout_area) AS avg_area,
       AVG(sun_201806211200_mean) AS avg_sun,
       AVG(noise_traffic_day) AS avg_noise
FROM rooms
GROUP BY layout_area_type
```

```python
result_df = pd.read_sql_query("""
    SELECT layout_area_type,
           AVG(layout_area) AS avg_area,
           AVG(sun_201806211200_mean) AS avg_sun,
           AVG(noise_traffic_day) AS avg_noise
    FROM rooms
    GROUP BY layout_area_type
""", conn)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
result_df.plot(kind='bar', x='layout_area_type', y='avg_sun', ax=axes[0], title='Avg Sun')
result_df.plot(kind='bar', x='layout_area_type', y='avg_noise', ax=axes[1], title='Avg Noise')
result_df.plot(kind='bar', x='layout_area_type', y='avg_area', ax=axes[2], title='Avg Area')

# plt.tight_layout()
plt.show()
```
